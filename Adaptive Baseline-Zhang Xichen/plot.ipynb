```python
import pickle
import matplotlib.pyplot as plt
def draw_plot(train_data, test_data):
    with open(train_data, 'rb') as f:
        data1 = pickle.load(f)
    with open(test_data, 'rb') as f:
        data2 = pickle.load(f)
    total_rewards = data1['total_rewards']
    final_rewards = data1['final_rewards']
    moving_average_rewards = data1['average_rewards']
    test_total_rewards = data2['total_rewards']
    test_final_rewards = data2['final_rewards']
    # step_used_list = data2['step_used_list']
    
    plt.figure(figsize=(16,8))

    plt.subplot(1, 2, 1)
    plt.title("Baseline training result")
    plt.plot(total_rewards, label='Total Rewards')
    plt.plot(moving_average_rewards, color='red', label='Average Rewards')
    plt.scatter(list(range(len(final_rewards))),final_rewards, color='green', label='Final Rewards', s=1)
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.ylim(-300, 300)
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.title("Baseline evaluation result")
    plt.hist(test_total_rewards, bins=50, color='green')
    plt.xlabel('Total Reward')
    plt.ylabel('Frequency')
    success_rate = test_final_rewards.count(100)/ 500
    plt.text(0.5, 0.5, f'Succesful landing rate: {success_rate}', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)

    plt.show()
```

```python
draw_plot('data/train_process/fix_random_train_data.pkl','data/test_process/fix_random_test_data.pkl')
draw_plot('data/train_process/luck_draw_train_data.pkl','data/test_process/luck_draw_test_data.pkl')
draw_plot('data/train_process/data_LunarLander_0.02_0.9_0.999.pkl','data/test_process/model_test_LunarLander_0.02_0.9_0.999.pth.pkl')
```
